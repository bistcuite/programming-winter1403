# مسئله اول

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 500)
```

```python
y =  0.5 * np.cos(2 * x) + np.sin(x)
```

```python
plt.plot(x, y)
plt.title('f(x) = 0.5 cos(2x) + sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
```

# مسئله دوم

```python
def f(x):
  return  0.5 * np.cos(2 * x) + np.sin(x)
```

```python
a = 0 # کران پایین
b = 3 # کران بالا
n = 1000000 # تعداد مستطیل ها
total = 0

dx = (b - a) / n
```

```python
for i in range(n):
  x = a + i * dx
  total += f(x) * dx

total
```
